```python
percent = 0.6
#graph_type = "ppm"

percentages = [0.6, 0.9]
graph_types = ['lfr', 'ppm']

start_step = 0.05
step_size = 0.05
end_steps = {"lfr": 0.5, "ppm": 0.9}


for percent in percentages:
    print(f"This is the check session for {percent}!!!")
    for graph_type in graph_types:
        end_step = end_steps[graph_type]
        MU = np.around(np.arange(start_step, end_step + 0.01, step_size), decimals=2)
        for mu in MU:
            for i in range(50):
                graphs_original = load_graph_only(mu, graph_type, "original", percent)
                graphs_sparse = load_graph_only(mu, graph_type, "sparse", percent)
                print(graphs_sparse[i].number_of_edges()/graphs_original[i].number_of_edges())
```
