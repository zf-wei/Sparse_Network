```python
import pickle
from clusim.clustering import Clustering
from sklearn.metrics import normalized_mutual_info_score
import clusim.sim as sim

def similarity_score(mu, graph_type, delete_type, embedding_method):
    """Process a specific mixing parameter (mu) to do community detection."""
    mu_str = f"{mu:.2f}"
    detected_dir = f'communitydetection_{delete_type}'
    raw_euclid_path = f'{detected_dir}/{graph_type}_{delete_type}_{embedding_method}_euclid_mu{mu_str}.pkl'
    raw_cosine_path = f'{detected_dir}/{graph_type}_{delete_type}_{embedding_method}_cosine_mu{mu_str}.pkl'
    with open(raw_euclid_path, 'rb') as file:
        euclid_membership = pickle.load(file)
    with open(raw_cosine_path, 'rb') as file:
        cosine_membership = pickle.load(file)
    _, original_membership = load_graph(mu, graph_type, "original")

    raw_score_mu = np.zeros((len(euclid_membership),4))
    for i in range(len(euclid_membership)):
        raw_score_mu[i,0] = normalized_mutual_info_score(original_membership[i], euclid_membership[i], average_method='arithmetic')
        raw_score_mu[i,1] = normalized_mutual_info_score(original_membership[i], cosine_membership[i], average_method='arithmetic')

        euclid_membership_clus = Clustering({j: [euclid_membership[i][j]] for j in range(len(euclid_membership[i]))})
        cosine_membership_clus = Clustering({j: [cosine_membership[i][j]] for j in range(len(cosine_membership[i]))})
        original_membership_clus = Clustering({j: [original_membership[i][j]] for j in range(len(original_membership[i]))})
        
        raw_score_mu[i,2] = sim.element_sim(original_membership_clus, euclid_membership_clus, alpha=0.9)
        raw_score_mu[i,3] = sim.element_sim(original_membership_clus, cosine_membership_clus, alpha=0.9)
    #print(raw_score_mu)
        
    os.makedirs(f'results_{delete_type}', exist_ok=True)
    raw_score_path = f'results_{delete_type}/{graph_type}_{delete_type}_{embedding_method}_raw_score_mu{mu_str}.pkl'
    with open(raw_score_path, 'wb') as file:
        pickle.dump(raw_score_mu, file)
    # print(f"RAW_SCORE for mu={mu_str} saved to {raw_score_path}")
```

```python
mu=0.05
graph_type = 'lfr'
delete_type = 'original'
embedding_method = "lap" # "lap" or "n2v"

similarity_score(mu, graph_type, delete_type, embedding_method)
```

```python
start_step = 0.05
step_size = 0.05
graph_type = "lfr"
end_steps = {"lfr": 0.5, "ppm": 0.9}
end_step = end_steps[graph_type]
MU = np.around(np.arange(start_step, end_step + 0.01, step_size), decimals=2)

graph_types = ['lfr', 'ppm']
delete_types = ['original', 'random', 'sparse']
embedding_method = 'lap'


for graph_type in graph_types:
    for delete_type in delete_types:
        end_step = end_steps[graph_type]
        MU = np.around(np.arange(start_step, end_step + 0.01, step_size), decimals=2)
        for mu in MU:
            similarity_score(mu, graph_type, delete_type, embedding_method)
            print(f"Work Done for {graph_type}, {delete_type}, mu={mu}")
```
